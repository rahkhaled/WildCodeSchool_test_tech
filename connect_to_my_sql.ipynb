```python
for i in range (len(df)) : 
    try:
        pos_lat  = (geocoder.geocode(df[1].loc[i] +" "+df[2].loc[i] +" "+df[3].loc[i])).latitude
        pos_long  = (geocoder.geocode(df[1].loc[i] +" "+df[2].loc[i] +" "+df[3].loc[i])).longitude
    except : 
        df.lat.loc[i] = "Nan"
        df.long.loc[i] = "Nan"
    else :
        df.lat.loc[i] = pos_lat
        df.long.loc[i] = pos_long  
df
```
